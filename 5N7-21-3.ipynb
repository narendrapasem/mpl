## Handling Missing Data

```python
import numpy as np
import pandas as pd
df=pd.read_csv("deepak sir dataset.csv")
df
```

```python
df.columns=['cl1','cl2','cl3','cl4','cl5','cl6','cl7','cl8','cl9']
```

```python
df
```

```python
columns=['cl1','cl2','cl3','cl4','cl5','cl6','cl7','cl8','cl9']
df[columns]=df[columns].replace(['0',0],np.nan)
```

```python
df
```

```python
cwithna=[features for features in df.columns if df[features].isnull().sum()>1]
print("missing columns:",len(cwithna))
```

```python
for feature in cwithna:
 print(feature,':',np.round(df[feature].isnull().mean()*100,2),"% missing values")
```

# Filling Method 5:median

```python
import numpy as np
import pandas as pd
df5=pd.read_csv("deepak sir dataset.csv")
df5.columns=['cl1','cl2','cl3','cl4','cl5','cl6','cl7','cl8','cl9']
df5[columns]=df5[columns].replace(['0',0],np.nan)
```

```python
columns=['cl1','cl2','cl3','cl4','cl5','cl6','cl7','cl8','cl9']
df5[columns]=df5[columns].replace(['0',0],np.nan)
```

```python
df5
```

```python
for feature in df5[columns]:
    df5[feature].median()
    
```

```python
for featu in df5[columns]:
    df5[featu].fillna(df5[featu].median(),inplace=True)
```

```python
df5.isna().sum()
```

```python
df5
```

```python
cwithna=[features for features in df5.columns if df5[features].isnull().sum()>1]
print("missing values",len(cwithna))
```

# Filling Method 6:mode

```python
import numpy as np
import pandas as pd
df6=pd.read_csv("deepak sir dataset.csv")
df6.columns=['cl1','cl2','cl3','cl4','cl5','cl6','cl7','cl8','cl9']
df6[columns]=df6[columns].replace(['0',0],np.nan)
df6
```

```python
columns=['cl1','cl2','cl3','cl4','cl5','cl6','cl7','cl8','cl9']
df6[columns]=df6[columns].replace(['0',0],np.nan)
```

```python
for features in df6[columns]:
    df6[features].mode()
```

```python
for featu in df6[columns]:
    df6[featu].fillna(df6[featu].mode()[0],inplace=True)
```

```python
df6.isna().sum()
```

```python
cwithna=[features for features in df6.columns if df6[features].isnull().sum()>1]
print('missing values',len(cwithna))
```
